```python
import numpy as np 
import matplotlib.pyplot as plt

def draw(filo: str, name: str):
    x = []
    y = []
    xmin = 4
    xmax = 5100000

    with open(filo, 'r') as filo:
        for i in filo.readlines():
            a, b = i.split('\t')
            a = a.replace(',', '.')
            b = b.replace('\n', '').replace(',', '.')
            x.append(float(a))
            y.append(float(b))
    
    figure, axes = plt.subplots()

    # plt.title("Зависимость напряжения на {name} от частоты".format(name=name))

    axes.set_xlabel("Сила тока, А")
    axes.set_ylabel("U, В")
    axes.semilogx(x)
    axes.set_xlim(xmin, xmax)
    axes.set_ylim(-1, 10)
    axes.grid()
    plt.figure(dpi=100)
    axes.scatter(x, y, c='black')

draw("gdm.txt", "GDM-8145")
draw("dt.txt", "DT838")
```
